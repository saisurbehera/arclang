```python
def find_border_and_inner(img: Image, find_border: bool = False) -> List[Tuple[Image, Point]]:
    border_color = img.majority_col()
    
    components = list_components(img)
    
    border_components = []
    inner_components = []
    
    for component, point in components:
        if component.majority_col() == border_color:
            border_components.append((component, point))
        else:
            inner_components.append((component, point))
    
    return border_components if find_border else inner_components
```
